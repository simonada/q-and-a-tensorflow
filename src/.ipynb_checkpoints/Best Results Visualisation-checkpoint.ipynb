```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.plotly as py
import matplotlib.patches as mpatches
```

# The visualized data reflects the best test prediction accuracy from three independent runs of 200 epochs

```python
# data to plot
n_groups = 3
f, (ax1, ax2, ax3) = plt.subplots(3, sharey=True, figsize=(6, 10))

 
# plot 1   
metric_accuracy_sso= (0.91, 0.86, 0.85)
metric_accuracy_all_sent= (0.68, 0.70, 0.63)

index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = ax1.bar(index, metric_accuracy_sso, bar_width,
                 alpha=opacity,
                 color='orange',
                 label='Supporting only')
 
rects2 = ax1.bar(index + bar_width, metric_accuracy_all_sent, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Full context')
 
ax1.set_xlabel('Task 5: Three argument relations')
ax1.set_ylabel('Accuracy')
ax1.set_xticks(index + bar_width)
ax1.set_xticklabels(('Low-level API', 'tf.keras', 'Eager Execuction'))
handles1, labels1 = ax1.get_legend_handles_labels()
ax1.legend(handles1, labels1)

# plot 2
metric_accuracy_sso= (0.70, 0.55, 0.51)
metric_accuracy_all_sent= (0.53, 0.52, 0.51)

rects3 = ax2.bar(index, metric_accuracy_sso, bar_width,
                 alpha=opacity,
                 color='orange',
                 label='Supporting only')
 
rects4 = ax2.bar(index + bar_width, metric_accuracy_all_sent, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Full context')
 
ax2.set_xlabel('Task 6: Yes/No questions')
ax2.set_ylabel('Accuracy')
ax2.set_xticks(index + bar_width)
ax2.set_xticklabels(('Low-level API', 'tf.keras', 'Eager Execution'))
handles2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(handles2, labels2)

# plot 3
metric_accuracy_sso= (0.73, 0.57, 0.57)
metric_accuracy_all_sent= (0.49, 0.44, 0.47)

rects5 = ax3.bar(index, metric_accuracy_sso, bar_width,
                 alpha=opacity,
                 color='orange',
                 label='Supporting only')
 
rects6 = ax3.bar(index + bar_width, metric_accuracy_all_sent, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Full context')
 
ax3.set_xlabel('Task 10: Indefinite knowledge')
ax3.set_ylabel('Accuracy')
ax3.set_xticks(index + bar_width)
ax3.set_xticklabels(('Low-level API', 'tf.keras', 'Eager Execution'))
handles3, labels3 = ax3.get_legend_handles_labels()
ax3.legend(handles3, labels3)

plt.tight_layout()
plt.show()
```
